```python
# Define the base directory
base_dir = 'H:\\CSIC\\DATA_PREPROCESSED_MICROSTATES'

# Walk through the directory
for root, dirs, files in os.walk(base_dir):
    for file in files:
        file_path = os.path.join(root, file)
        # Check if the file is empty (0 bytes)
        try:
            if os.path.getsize(file_path) == 0:
                segment = file_path.split("\\")[-2]
                patient_id = file_path.split("\\")[-1].split('_')[0]
                u.move_error_files(patient_id)
                #print(f"Empty file found: {file} at {segment}")
        except FileNotFoundError:
            continue
```

```python
# Define the base directory
base_dir = 'H:\\CSIC\\ERRORS'
empty = []
# Walk through the directory
for root, dirs, files in os.walk(base_dir):
    for file in files:
        file_path = os.path.join(root, file)
        # Check if the file is empty (0 bytes)
        try:
            if os.path.getsize(file_path) == 0:
                segment = file_path.split("\\")[-2]
                patient_id = file_path.split("\\")[-1].split('_')[0]
                transition = file_path.split("\\")[-1].split('_')[1].split(".")[0]
                empty.append((segment, patient_id, transition))
        except FileNotFoundError:
            continue
```

```python
from collections import defaultdict

# Initialize the main dictionary with defaultdict to handle nested dictionaries
segment_dict = defaultdict(lambda: defaultdict(list))

# Iterate over the data
for segment, patient_id, transition in empty:
    # Append the transition to the list for the specific patient in the segment
    segment_dict[segment][patient_id].append(transition)

# Convert defaultdict back to regular dict if needed
segment_dict = {segment: dict(patients) for segment, patients in segment_dict.items()}

# Print the result
for segment, patients in segment_dict.items():
    print(f"\n{segment}:")
    for patient_id, transitions in patients.items():
        print(f"- {patient_id}: {transitions}")
```
